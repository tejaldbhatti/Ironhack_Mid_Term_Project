# Hypothesis Testing

```python
ave_FEE1 = FEE["january"].mean()
print(ave_FEE1)
ave_FEE2 = FEE["june"].mean()
print(ave_FEE2)
```

```python
from scipy.stats import ttest_1samp

# one-sided test:
#H0: average FEE in month of JAnuary is = average FEE in month of June 
#H1:  average FEE in month of JAnuary is <> average FEE in month of June

stat, pval = ttest_1samp(FEE["january"],1.57)
print('pvalue for the one-tailed test is ', pval)
```

```python
country_and_reason = pd.read_csv("India-Tourism-Statistics-2019_region-and-reason-Copy1.csv")
```

```python
country_and_reason.columns = country_and_reason.columns.str.replace(" ","_").str.lower()
country_and_reason
```

```python
country_and_reason.isna().sum()
```

```python
region_reason = country_and_reason[country_and_reason["country_of_nationality"]=="Total"].copy()
region_reason["business_and_professional(%)"] = region_reason["business_and_professional(%)"]*region_reason["arrivals_(in_numbers)"]/100*10**-5
region_reason["leisure_holiday_and_recreation(%)"] = region_reason["leisure_holiday_and_recreation(%)"]*region_reason["arrivals_(in_numbers)"]/100*10**-5
region_reason["medical(%)"] = region_reason["medical(%)"]*region_reason["arrivals_(in_numbers)"]/100*10**-5
region_reason["indian_diaspora(%)"] = region_reason["indian_diaspora(%)"]*region_reason["arrivals_(in_numbers)"]/100*10**-5
region_reason["others(%)"] = region_reason["others(%)"]*region_reason["arrivals_(in_numbers)"]/100*10**-5
region_reason

```

```python
region_reason.drop(["country_of_nationality","arrivals_(in_numbers)"],axis =1,inplace = True)
```

```python
region_reason
```

```python
region_reason.rename(columns={
    'business_and_professional()':'Business and Professional',
    'leisure_holiday_and_recreation()':'Leisure Holiday and Recreation',
    'medical()':'Medical',
    'indian_diaspora()':'Indian Diaspora',
    'others()':'Others'
},inplace=True)

# Resetting index
region_reason.set_index('region',inplace=True)
```

```python
region_reason.to_csv("Region_Reason.csv")
```

```python
plt.figure(figsize=(100,80))
region_reason.plot()
plt.show()
```

```python
countrywise=country_and_reason[country_and_reason['country_of_nationality'].isin(['Total','Grand Total','Not Classified Elsewhere','OTHERS'])==False].copy()
# Sorting w.r.t Arrivals in descending order
countrywise.sort_values(by=["arrivals_(in_numbers)"],ascending=False, inplace = True)
# fetching 5 countries with top arrivals
countrywise=countrywise.iloc[0:5,:]
# dropping column 'Region'
countrywise.drop(["region"],axis=1,inplace=True)
```

```python
countrywise
```

```python

countrywise["business_and_professional(%)"] = countrywise["business_and_professional(%)"]*countrywise["arrivals_(in_numbers)"]/100*10**-3
countrywise["leisure_holiday_and_recreation(%)"] = countrywise["leisure_holiday_and_recreation(%)"]*countrywise["arrivals_(in_numbers)"]/100*10**-3
countrywise["medical(%)"] = countrywise["medical(%)"]*countrywise["arrivals_(in_numbers)"]/100*10**-3
countrywise["indian_diaspora(%)"] = countrywise["indian_diaspora(%)"]*countrywise["arrivals_(in_numbers)"]/100*10**-3
countrywise["others(%)"] = countrywise["others(%)"]*countrywise["arrivals_(in_numbers)"]/100*10**-3
countrywise
countrywise.drop(["arrivals_(in_numbers)"],axis=1,inplace=True)
df=pd.melt(countrywise,["country_of_nationality"]).rename(columns={"variable":"Reason of Visit"})
```

```python
df
```

```python
# plotting bar (multicolumn) graph
plt.figure(figsize=(20,8))
# country v / s reason barplot
sns.barplot(x ="country_of_nationality",y='value', hue="Reason of Visit",data = df, palette='Paired')
```
