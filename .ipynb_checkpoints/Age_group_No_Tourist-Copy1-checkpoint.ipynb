```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
age_group=pd.read_csv("India-Tourism-Statistics-2001-2019-agegroup(1).csv")
```

```python
age_group.head()
```

# Data Cleaning


```python
def col_correction(a):
    x = a.lower()
    if "% distribution by age-group (in year) - " in x:
        return  (x.replace('% distribution by age-group (in year) - ',""))
    else:
        return x
age_group.columns = list(map(col_correction,age_group))    
```

```python
age_group.to_csv("AGE GROUP With total tourist.csv")
```

```python
age_group
```

# check for nulls

```python
age_group.isnull().sum()
```

```python
# dropping columns
age_group.drop(['ftas','not reported'],axis=1 ,inplace=True)
# Combining columns w.r.t 'Year' into a single column 'value' and column 'variable' will distinguish these values 
df=pd.melt(age_group,['year']).rename(columns={"variable":"Age Group"})
```

```python
df
```

```python
df.to_csv("Age group year wise.csv")
```

```python
plt.figure(figsize=(20,8))

sns.set_context("talk")

g=sns.lineplot(x='year', y='value', hue='Age Group', data=df)
g.set(xlabel="Year",ylabel="Number of tourists (in thousands)", title="Tourists segregation based on age [2001-2019]")
plt.xticks(
    range(2001,2020),
    rotation=45,
#     horizontalalignment='right',
    fontweight='light'
)

# display the chart
plt.plot()
```

# Hypothesis Testing

```python
ave_mid_age = age_group["45-54"].mean()
print(ave_mid_age)
ave_65_and_above = age_group["65 and above"].mean()
print(ave_65_and_above)
```

```python
from scipy.stats import ttest_1samp

# two-sided test:
#H0: average number of visitors of age group 45-54 >= average number of visitors of age group 65 and above
#H1:  average number of visitors of age group 45-54 < average number of visitors of age group 65 and above

stat, pval = ttest_1samp(age_group["45-54"],8.215)
print('pvalue for the two-tailed test is ', pval/2)
print('stat for the two-tailed test is ', stat)
```

```python
#P value < 0.05 we can reject equality but stat value is positive
#so we can not reject > sign . So overall we can not reject null hyposthesis
```
