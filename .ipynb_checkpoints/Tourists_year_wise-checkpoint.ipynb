# Data Cleaning 

```python
def column_name_Correction(x):
    b = x.lower()
    a = b.replace(" ","_")
    if "1st quarter" in x:
        d = a.replace("1st quarter","Q1").split("_(")[0]
    elif "2nd quarter" in x: 
        d = a.replace("2nd quarter","Q2").split("_(")[0]
    elif "3rd quarter" in x: 
        d = a.replace("3rd quarter","Q3").split("_(")[0]  
    else:
        d = a.replace("4th quarter","Q4").split("_(")[0]    
    return d  
c = list(map(column_name_Correction,data_Population3.columns))
```

```python
b = data_Population3.columns 
keys = b
values = c

# Create dictionary using zip() function
data_dict = dict(zip(keys, values))
data_Population3.rename(columns = data_dict,inplace = True)
data_Population3
```

# Hypothesis Testing

```python
ave_Q4 = tourist_Quartely_year_wise["quarter_4"].mean()
print(ave_Q4)
ave_Q2 = tourist_Quartely_year_wise["quarter_2"].mean()
print(ave_Q2)
```

```python
from scipy.stats import ttest_1samp

# Two-sided test:
#H0: average number of visitors in Quarter 4 >= average number of visitors in Quarter 2
#H1:  average number of visitors in Quarter 4 <= average number of visitors in Quarter 2

stat, pval = ttest_1samp(tourist_Quartely_year_wise["quarter_4"],19.068)
print('pvalue for the two-tailed test is ', pval/2)
print('stat for the two-tailed test is ', stat)
```

```python
#Here p value is <0.005 so we can reject the = sign but our stat value is positive so we can not reject the null hypothesis
```
